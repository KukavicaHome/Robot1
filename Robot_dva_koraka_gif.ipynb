```python

# 1. Učitaj potrebne biblioteke
from PIL import Image
from IPython.display import Image as IPImage, display
import os

# 2. Omogući upload slika
from google.colab import files

print("Molimo uploadujte dvije slike: jedna s desnim korakom, jedna s lijevim.")
uploaded = files.upload()

# 3. Pronađi imena fajlova koje si uploadovao
uploaded_files = list(uploaded.keys())
if len(uploaded_files) < 2:
    raise ValueError("Potrebne su dvije slike za simulaciju hoda.")

# 4. Učitaj slike
img1 = Image.open(uploaded_files[0])
img2 = Image.open(uploaded_files[1])

# 5. Izjednači veličinu ako je potrebno
img2 = img2.resize(img1.size)

# 6. Napravi GIF
output_name = "walking_cycle.gif"
img1.save(
    output_name,
    save_all=True,
    append_images=[img2],
    duration=400,
    loop=0
)

# 7. Prikaži i ponudi za download
print("GIF je napravljen.")
display(IPImage(filename=output_name))
files.download(output_name)
```
